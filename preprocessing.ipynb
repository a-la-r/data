```python
import pandas as pd
```

```python
df_kak_wr = pd.read_csv('df_kak_wr.csv')
df_slovno_wr = pd.read_csv('df_slovno_wr.csv')
df_aki_wr = pd.read_csv('df_aki_wr.csv')
df_ala_wr = pd.read_csv('df_ala_wr.csv')
df_budto_wr = pd.read_csv('df_budto_wr.csv')
df_podobno_wr = pd.read_csv('df_podobno_wr.csv')
```

```python
df_kak_250 = df_kak_wr.sample(n=250)
df_slovno_250 = df_slovno_wr.sample(n=250)
df_aki_250 = df_aki_wr.sample(n=250)
df_ala_250 = df_ala_wr.sample(n=250)
df_budto_250 = df_budto_wr.sample(n=250)
df_podobno_250 = df_podobno_wr.sample(n=250)
```

```python
frames = [df_kak_250, df_slovno_250, df_aki_250, df_ala_250, df_budto_250, df_podobno_250]
df_250_wr = pd.concat(frames) 
```

```python
df_250_wr = df_250_wr.reset_index(drop=True)
```

```python
df_250_wr
```

```python
for i, text in enumerate(df_250_wr['Sphere']):
    df_250_wr['Sphere'][i] = text.split(',')[0]
```

```python
for i, text in enumerate(df_250_wr['Sphere']):
    if df_250_wr['Sphere'][i] == 'бытовая':
        df_250_wr['Sphere'][i] = 'нехудожественная'
    elif df_250_wr['Sphere'][i] == 'художественная проза':
        df_250_wr['Sphere'][i] = 'художественная'
    elif df_250_wr['Sphere'][i] == 'публицистика':
        df_250_wr['Sphere'][i] = 'нехудожественная'

        
```

```python
for i, text in enumerate(df_250_wr['Type']):
     df_250_wr['Type'][i] = text.split(',')[0]
```

```python
for i, text in enumerate(df_250_wr['Center']):
    df_250_wr['Center'][i] = text.lower()
```

```python
for i, text in enumerate(df_250_wr['Left context']):
    try:
        df_250_wr['Left context'][i] = text.lower()
    except:
        continue
```

```python
for i, text in enumerate(df_250_wr['Right context']):
    try:
        df_250_wr['Right context'][i] = text.lower()
    except:
        continue
```

```python
for i, text in enumerate(df_250_wr['Created']):
    if len(text) == 4:
        text = list(text)
        text[3] = str(0) 
        text = "".join(text)
        #print(text)
        df_250_wr['Created'][i] = text
    else:
        text = list(text)
        text[3] = str(0)
        text = text[0:4]
        text = "".join(text)
        df_250_wr['Created'][i] = text
```
