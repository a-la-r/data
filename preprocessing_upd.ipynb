```python
science = ['монография', 'учебное пособие', 'учебник', 'лекция']
business = ['отчет', 'записка докладная', 'устав', 'послание']
bookish = ['фельетон', 'репортаж', 'статья', 'очерк', 'дневник', 'мемуары', 'доклад', 'эссе', 'предисловие', 'рецензия', 
           'отзыв', 'анонс', 'катехизис', 'трактат', 'обзор', 'блог', 'комментарий', 'интервью', 'заметка', 'записки', 
           'биография', 'гороскоп', 'информационное сообщение']
fiction = ['рассказ', 'роман', 'повесть', 'сказ', 'сказка', 'пьеса', 'миниатюра', 'анекдот', 'легенда', 'цикл']
oral = ['беседа', 'письмо личное', 'речь', 'слово', 'проповедь', 'поучение', 'форум']

for i, text in enumerate(df_250_wr['Type']):
    if text in science:
        df_250_wr['Type'][i] = 'научный'
    elif text in business:
        df_250_wr['Type'][i] = 'деловой'
    elif text in bookish:
        df_250_wr['Type'][i] = 'публицистический'
    elif text in fiction:
        df_250_wr['Type'][i] = 'художественный'
    elif text in oral: 
        df_250_wr['Type'][i] = 'разговорный'
```

1) ADJ - ADJF, ADJS, COMP
2) AUX - PREP, CONJ, PRCL
3) TRSH - nan, UNKN, PNCT, INTJ, PRED
4) LATN
5) VERB - INFN, VERB
6) NOUN
7) PRT_GRND - PRTS, PRTF, GRND
8) NUMR - NUMB, NUMR
9) ADVB
10) NPRO

```python
adj = ['ADJF', 'ADJS', 'COMP']
aux = ['PREP', 'CONJ', 'PRCL']
trsh = ['NaN', 'UNKN', 'PNCT', 'INTJ', 'PRED']
verb = ['INFN', 'VERB']
prt_grnd = ['PRTS', 'PRTF', 'GRND']
numr = ['NUMB', 'NUMR']

for i, text in enumerate(df_250_wr['Left_context_POS']):
    if text in adj:
        df_250_wr['Left_context_POS'][i] = 'ADJ'
    elif text in aux:
        df_250_wr['Left_context_POS'][i] = 'AUX'
    elif text in trsh:
        df_250_wr['Left_context_POS'][i] = 'TRSH'
    elif text in verb:
        df_250_wr['Left_context_POS'][i] = 'VERB'
    elif text in prt_grnd:
        df_250_wr['Left_context_POS'][i] = 'PRT_GRND'
    elif text in numr:
        df_250_wr['Left_context_POS'][i] = 'NUMR'
```

```python
for i, text in enumerate(df_250_wr['Right_context_POS']):
    if text in adj:
        df_250_wr['Right_context_POS'][i] = 'ADJ'
    elif text in aux:
        df_250_wr['Right_context_POS'][i] = 'AUX'
    elif text in trsh:
        df_250_wr['Right_context_POS'][i] = 'TRSH'
    elif text in verb:
        df_250_wr['Right_context_POS'][i] = 'VERB'
    elif text in prt_grnd:
        df_250_wr['Right_context_POS'][i] = 'PRT_GRND'
    elif text in numr:
        df_250_wr['Right_context_POS'][i] = 'NUMR'
```

```python
df_250_wr['Left_context_POS'].replace(np.nan, 'TRSH', inplace=True)
df_250_wr['Right_context_POS'].replace(np.nan, 'TRSH', inplace=True)
```
